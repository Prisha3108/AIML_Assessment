# Detecting Outliers

```python
import numpy as np

# Define a function to identify outliers based on IQR
def detect_outliers_iqr(data, column):
    Q1 = np.percentile(data[column], 25)  # 1st quartile
    Q3 = np.percentile(data[column], 75)  # 3rd quartile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR  # Lower bound for outliers
    upper_bound = Q3 + 1.5 * IQR  # Upper bound for outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect outliers in the 'Sourcing Cost' column
outliers_sourcing_cost = detect_outliers_iqr(train_data, 'Sourcing Cost')

# Print the outliers
print("Outliers in Sourcing Cost:")
print(outliers_sourcing_cost)
```

# Removing Outliers

```python
# Define a function to remove outliers based on IQR
def remove_outliers_iqr(data, column):
    Q1 = np.percentile(data[column], 25)  # 1st quartile
    Q3 = np.percentile(data[column], 75)  # 3rd quartile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR  # Lower bound for outliers
    upper_bound = Q3 + 1.5 * IQR  # Upper bound for outliers
    # Filter out rows where the column values are within the bounds
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Remove outliers from the training data
train_data_no_outliers = remove_outliers_iqr(train_data, 'Sourcing Cost')

# Visualize the distribution of sourcing costs after removing outliers
plt.figure(figsize=(10, 5))
sns.histplot(data=train_data_no_outliers, x="Sourcing Cost", kde=True)
plt.title("Distribution of Sourcing Costs (Outliers Removed)")
plt.xlabel("Sourcing Cost")
plt.ylabel("Count")
plt.show()
```

Gradient Boosting:
Gradient Boosting is another ensemble method that builds decision trees sequentially, each one correcting the errors of its predecessor.
